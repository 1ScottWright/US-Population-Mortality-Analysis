# EDA on potential target variables

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
plt.style.use('wright')
```

```python
df = pd.read_csv('../data/features.csv', index_col=1)
df = df.drop(['Unnamed: 0','Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6'], axis=1)
```

```python
df[59:80]
```

#### Plot a graph of each column

```python
col = list(df.columns)
colors = ['#66BF6A', '#3DC4D7', '#FBC94D', '#F17B33', '#002A5C', '#6D6F72', '#df9fdf']
n = 0

for each in col:
    
    plt.plot(df[each].dropna(), color = colors[n % 6])
    plt.title("Variable: "+each)
    plt.show();
    n+=1
```

#### Plotting notes:
> GDP for 2018 is incomplete and should not be used.  Currently, it is not part of the analysis since only variables up to 2015 will be used.   

> The 10 year treasury rate item looks suspect -- the value is near 30% around 1980.  This also requires follow-up.
  
> The overweight measure seems counterintuitive and requires follow-up

#### Follow-up notes:
>  The 10 year treasury is the return over the year if invested in 10 year treasuries as of Jan 1.  Thus, a decrease in interest rates during the year could dramatically increase the yiled beyond the nominal rate.  

> The category `overweight` do not include `obese` or `severely_obese`.  As more of the pupulation shifted to `obese` and `severely_obese`, this increase was not always accompanied by a shift from normal to `overweight`.
